I scraped all the lyrics from the songs on the 'What Could Possibly Go Wrong' (2020) album by Dominic Fike, using the Genius API, and turned them into TXT files. Moreover, I put those files in the data directory.

```python
# Install the LyricsGenius library using John W. Miller's 
!pip install git+https://github.com/johnwmillr/LyricsGenius.git

# My client access token from the Genius API
client_access_token = "w_aWx0rdBwUR_qT_3Pbmm_TCe65XKDjsZ-ebGbC-oCw2qFjUYesiQl6dbKGRc0R5"

# Import libraries
import lyricsgenius, pathlib, os

# Assign my client access token to the LyricsGenius variable
LyricsGenius = lyricsgenius.Genius(client_access_token)

# All songs from the 'What Could Possibly Go Wrong' album by Dominic Fike
songs = ["Come Here", "Double Negative (Skeleton Milkshake)", "Cancel Me", "10x Stronger", "Good Game", "Why", "Chicken Tenders", "Whats For Dinner?", "Vampire", "Superstar Sh*t", "Politics & Violence", "Joe Blazey", "Wurli", "Florida"]

# Absolute path for the directory where the TXT files will be stored
directory = r"C:/Users/camer/OneDrive/Documents/University/MA Digital Humanities/Collecting Data/Assignment_2/Assignment_2/data"

# Iterate over the songs in the songs list
for song in songs:
    # Search for Dominic Fike song with relevant title
    specific_song = LyricsGenius.search_song("Dominic Fike", song)
    # Save song lyrics as TXT file
    specific_song.save_lyrics(extension='txt')

# Save TXT files in list
lyrics = [l for l in os.listdir() if '.txt' in l.lower()]

# Create a folder named data
os.mkdir('data')

# Iterate over TXT files
for lyric in lyrics:
    # Createnew path for TXT files
    new_path = 'data/' + lyric
    # Assign TXT files to data directory
    pathlib.Path(lyric).rename(new_path)
```
