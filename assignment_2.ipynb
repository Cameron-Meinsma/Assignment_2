# 2. Preprocessing

```python
# Import SpaCy and os
import spacy, os

# Install English language model
!spacy download en_core_web_sm

# Load SpaCy visualizer
from spacy import displacy
from IPython.display import display, HTML

# Import pandas DataFrame packages
import pandas as pd
pd.options.mode.chained_assignment = None

# Import graphing package
import plotly.express as px
```

```python
# Create empty lists for file names and contents
lyrics = []
file_names = []

# iterate through each file in folder
for _file_name in os.listdir('data'):
    # Look for only TXT files
    if _file_name.endswith('.txt'):
        # Append contents of each TXT file to text list
        lyrics.append(open('data' + '/' + _file_name, 'r', encoding='utf-8').read())
        # Append name of each file to file_names list
        file_names.append(_file_name)
```

```python
# Create dictionary object associating each file ame with its text
d = {'Filename':file_names, 'Document':lyrics}

# Turn dictionary into a dataframe
lyrics_df = pd.DataFrame(d)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

```python
# Remove extra spaces from the lyrics
lyrics_df['Text'] = lyrics_df['Document'].str.replace(r'\s+', ' ', regex=True).str.strip()
# Remove song parts from the lyrics
lyrics_df['Text'] = lyrics_df['Text'].str.replace(r'\[[a-zA-Z]+\]', '', regex=True).str.strip()

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

```python
# Remove .txt from the filenames
lyrics_df['Filename'] = lyrics_df['Filename'].str.replace('.txt', '', regex=True)

# Rename column from Filename to Song Title
lyrics_df.rename(columns={"Filename": "Song Title"}, inplace=True)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

# 3. Annotating

```python
# Load nlp pipeline
nlp = spacy.load('en_core_web_sm')

# Check what functions it performs
print(nlp.pipe_names)
```

```python
# Define a function that runs the nlp pipeline on any given input text
def process_text(text):
    return nlp(text)

# Apply the function to the "Text" column
lyrics_df['Doc'] = lyrics_df['Text'].apply(process_text)
```

```python
# Define a function that runs the nlp pipeline on any given input text
def process_text(text):
    return nlp(text)
```

```python
# Apply the function to the "Text" column, so that the nlp pipeline is called on every lyrics
lyrics_df['Doc'] = lyrics_df['Text'].apply(process_text)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

### 3.1 Tokenization

```python
# Define a fucntion to retrieve tokens from a doc object
def get_token(doc):
    return [(token.text) for token in doc]
```

```python
# Run the token retrieval function on the doc objects in the dataframe
lyrics_df['Tokens'] = lyrics_df['Doc'].apply(get_token)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

### 3.2 Lemmatization

```python
# Define a function to retrieve lemmas from a doc object
def get_lemma(doc):
    return [(token.lemma_) for token in doc]
```

```python
# Run lemma retrieval function on the doc objects in the dataframe
lyrics_df['Lemmas'] = lyrics_df['Doc'].apply(get_lemma)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

### 3.3 Parts-of-Speech Tagging

```python
# Define a function to retrieve lemmas from a doc object
def get_pos(doc):
    # Return the coars- and fine-grained part of speech text for each token in the doc
    return [(token.pos_, token.tag_) for token in doc]

# Define a function to retrieve parts of speech from a doc object
lyrics_df['Parts-of-speech'] = lyrics_df['Doc'].apply(get_pos)

# Create a list of part of speech tags
list(lyrics_df['Parts-of-speech'])

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```
