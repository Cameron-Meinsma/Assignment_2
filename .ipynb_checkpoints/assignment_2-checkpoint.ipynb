```python
# Install the LyricsGenius library using John W. Miller's adapted version
!pip install git+https://github.com/johnwmillr/LyricsGenius.git

# My client access token from the Genius API
client_access_token = "w_aWx0rdBwUR_qT_3Pbmm_TCe65XKDjsZ-ebGbC-oCw2qFjUYesiQl6dbKGRc0R5"

# Import libraries
import lyricsgenius, pathlib, os

# Assign my client access token to the LyricsGenius variable
LyricsGenius = lyricsgenius.Genius(client_access_token)

# All songs from the 'What Could Possibly Go Wrong' album by Dominic Fike
songs = ["Come Here", "Double Negative (Skeleton Milkshake)", "Cancel Me", "10x Stronger", "Good Game", "Why", "Chicken Tenders", "Whats For Dinner?", "Vampire", "Superstar Sh*t", "Politics & Violence", "Joe Blazey", "Wurli", "Florida"]

# Absolute path for the directory where the TXT files will be stored
directory = r"C:/Users/camer/OneDrive/Documents/University/MA Digital Humanities/Collecting Data/Assignment_2/Assignment_2/data"

# Iterate over the songs in the songs list
for song in songs:
    # Search for Dominic Fike song with relevant title
    specific_song = LyricsGenius.search_song("Dominic Fike", song)
    # Save song lyrics as TXT file
    specific_song.save_lyrics(extension='txt')

# Save TXT files in a list
lyrics = [l for l in os.listdir() if '.txt' in l.lower()]

# Create a directory named data
os.mkdir('data')

# Iterate over TXT files
for lyric in lyrics:
    # Createnew path for TXT files
    new_path = 'data/' + lyric
    # Assign TXT files to data directory
    pathlib.Path(lyric).rename(new_path)

# Change to data directory
os.chdir('C:/Users/camer/OneDrive/Documents/University/MA Digital Humanities/Collecting Data/Assignment_2/Assignment_2/data')

# Iterate over the TXT files
for file in lyrics:
    # Check if the file contains lyrics
    if file.startswith("saved_song_lyrics_dominic_fike_"):
        # Rename the file to only contain the title of the song
        os.rename(file, file[31:])

# Change to Assignment_2 directory
os.chdir('C:/Users/camer/OneDrive/Documents/University/MA Digital Humanities/Collecting Data/Assignment_2/Assignment_2')
```

# 2. Preprocessing

```python
# Import SpaCy and os
import spacy, os

# Install English language model
!spacy download en_core_web_sm

# Load SpaCy visualizer
from spacy import displacy
from IPython.display import display, HTML

# Import pandas DataFrame packages
import pandas as pd
pd.options.mode.chained_assignment = None

# Import graphing package
import plotly.express as px
```

```python
# Create empty lists for file names and contents
lyrics = []
file_names = []

# iterate through each file in folder
for _file_name in os.listdir('data'):
    # Look for only TXT files
    if _file_name.endswith('.txt'):
        # Append contents of each TXT file to text list
        lyrics.append(open('data' + '/' + _file_name, 'r', encoding='utf-8').read())
        # Append name of each file to file_names list
        file_names.append(_file_name)
```

```python
# Create dictionary object associating each file ame with its text
d = {'Filename':file_names, 'Lyrics':lyrics}

# Turn dictionary into a dataframe
lyrics_df = pd.DataFrame(d)

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```

```python
# Remove extra spaces from papers
lyrics_df['Lyrics'] = lyrics_df['Lyrics'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Show the head of the lyrics_df dataframe
lyrics_df.head()
```
